```python
def find_d(value):

    #d = my_string

    #folder_dir = ############f"C:/Users/garapati.koushik/Downloads/Project/{d}"#####################
    
    imge1 = []
    imge = []
    df = tppit[tppit["PSMIPSMRFNUM"]==value]
    imge1 = df["imgpath"].unique()
    
    x=range(20)
    count=0
    for i in x:
        a = random.choice(imge1)
        countpic=imge.count(a)
        if countpic < 1:
            imge.append(a)
        elif count>10:
            imge.append(a)
        count = count + 1
    return imge
        
```
